```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import numpy as np
```

```python
#Create a list for id of students
id =range(102300,1000000,1)

#List for info for each student
student_info=["id","name","school","district","divison","score","status"]

#List of all subjects
student_subjects=["arabic","first_lang","second_lang","math_1","history","geo",
                 "philo","psycho","chemistry","bio","geology","math_2","pyhsics"]
                 
student_secondary=["religion","national","stats"]

#List of subject of each divison
student_math=   ["arabic","first_lang","second_lang","chemistry","pyhsics","math_1","math_2"]
student_science=["arabic","first_lang","second_lang","chemistry","pyhsics","bio","geology"]
student_arts=   ["arabic","first_lang","second_lang","history","geo","philo","psycho"]

#Create a DataFrame to save the output of web scraping
students=pd.DataFrame(np.nan,index=id,columns=student_info+student_subjects+student_secondary)
students
```

```python
r=[]
t=time.time()
for i in range(30):
    #Scrap webpage for each student
    URL = "https://shbabbek.com/natega/"+ str(id[i])
    r.append(requests.get(URL))
    print(i)
    time.sleep(0.01)
print((time.time()-t)/30)    
```

```python
t=time.time()
for i in range(30):
    soup = BeautifulSoup(r[i].content, "lxml")
    if len(soup.find_all("tbody")) > 4:
        for j in range(13):
            x=soup.find_all("tbody")[4].find_all("tr")[j].find_all("td")[1].get_text().strip()
            if x!= 'غير مقرر':
                students.at[id[i],student_subjects[j]]=int(x)
        for j in range(3):
            x=soup.find_all("tbody")[4].find_all("tr")[j+14].find_all("td")[1].get_text().strip()
            if x!= 'غير مقرر':
                students.at[id[i],student_secondary[j]]=int(x)
        #for j in range(6):
            students.loc[id[i],student_info[j]]=soup.find_all("tbody")[3].find_all("tr")[j].find_all("td")[1].get_text().strip()
        students.loc[id[i],"status"]=soup.find_all("tbody")[3].find_all("tr")[8].find_all("td")[1].get_text().strip()
    print(i)
print((time.time()-t)/30)
```
